#### Question 1 - Read in the data set and print the % empty values for each column. Are there any columns with missing values?

```python
def data_reader(data_path):
    try:
        with open(data_path, 'r') as inF:
            print("File Successfully read")
            data = pd.read_csv(inF)
            print("The shape is {}".format(data.shape))
            return data;
        
    except IOError:
        print("Could not read file")
        return None
    
    
        
data_dir = 'C:\\Users\\Michael\\Desktop\\301gits\\HW6\\groundhog.csv'
```

```python
my = data_reader(data_dir)
```

```python
my.head()
```

```python
def get_percent_nans(df, column_name):
    nans = [n for n in df[column_name].values if pd.isnull(n)]
    percent = (len(nans)/len(df[column_name].values))*100
    return percent

for col in my.columns:
    print('The \"{}\" column has {:.2f}% empty values'.format(col,get_percent_nans(my,col)))
```

```python
column_list = ['February Average Temperature',
'February Average Temperature (Northeast)',
'February Average Temperature (Midwest)',
'February Average Temperature (Pennsylvania)',
'March Average Temperature',
'March Average Temperature (Northeast)',
'March Average Temperature (Midwest)',
'March Average Temperature (Pennsylvania)']

data_list = []
for col,data in my_copy.iteritems(): 
    if col in data_list:
        data_list.append((col,data))
        
for col, data in data_list:
    col_mean = data.mean()
    new_data = []
    for row_val in data:
        if pd.null(row_val):
            new_data.append(col_mean)
        else:
            new_data.append(row_val)
    my_copy[col] = new_data
    
            
my_copy
```
